```r
run.exp.D1 = function(){
    n = 10
    X = runif(n, -20, 20)

    # replace all values values greater than 6 with 6
    X = map_dbl(X, function(x) ifelse(x > 6, 6, x)) 
    m = mean(X)
    s = sd(X)
    t =  m/(s/sqrt(n))
}


run.exp.D2 = function(){
    n = 10
    X = runif(n, -20, 20)
    
    m = mean(X)
    s = sd(X)
    t =  m/(s/sqrt(n))
}



```

```r
future::plan("multiprocess") # Set to run in parallel
no_of_exps = 100000 # Set the number of experiments to run

D1_exps <- furrr::future_map_dbl(1:no_of_exps, function(x) 
  run.exp.D1())                                  

D2_exps <- furrr::future_map_dbl(1:no_of_exps, function(x) 
  run.exp.D2())
```

```r
tibble(D1 = D1_exps, D2 = D2_exps) %>% 
pivot_longer(cols = c("D1","D2"), values_to = "x") %>%
ggplot(aes(x = x,fill = name)) + 
geom_histogram(alpha = .5, bins = 50, na.rm = TRUE) + xlim(c(-4,4)) +
scale_fill_manual(values = c("D1" = "darkred","D2" = "darkblue"), 
                   labels = c("D1" = "Device D₁", "D2" = "Device D₂"), 
                   name = NULL) + 
theme_minimal() + theme(legend.position = "top") + 
labs(y = "number of experiments", x = "t") %>% suppressWarnings()
```

```r

do.flips = function(n_flips = 10){
    pr_heads = .5
    tibble(n = 10) %>% mutate(h = sum(purrr::rbernoulli(n,pr_heads)),t = n - h)
}

```

```r
all_flips = furrr::future_map_dfr(1:10000, function(x) do.flips(10))
```

```r
all_flips %>% add_count(name = "total")  %>% 
group_by(h,total) %>% summarise(d_heads = n()) %>%
mutate(d_heads = d_heads/total) %>%
ggplot(aes(x = h, y = d_heads)) + geom_point(size = 3, colour = "darkblue") + 
theme_minimal() + 
scale_x_continuous(name = "number of heads", limits = c(0,10), breaks = seq(0,10,2)) +
scale_y_continuous(name = "Relative occurrence")
```
