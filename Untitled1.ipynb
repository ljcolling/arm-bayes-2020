# Building more complex models

```r
library(greta)
library(tensorflow)
require(logspline)
require(tidyverse)
```

```r
set.seed(348)

x = rnorm(10, 12, 12)
y = rnorm(10, 32, 12)



zx = (x-mean(x))/sd(x)
zy = (y-mean(x))/sd(x)

delta = cauchy(0,1)
mu = cauchy(0,1)
sigma = cauchy(0,1, truncation = c(0,Inf))

alpha = delta*sigma

mux = mu + alpha/2
muy = mu - alpha/2

distribution(zx) = normal(mux,sigma)
distribution(zy) = normal(muy,sigma)

m = model(delta)

draws = mcmc(m, n_samples = 20000, chains = 4, n_cores = 6, verbose = F)

posteriorDelta = draws[[1]][,1]

fitPost = logspline(posteriorDelta)


BF_SD = dcauchy(0,0,1) / dlogspline(0,fitPost)

BF_pkg = BayesFactor::ttestBF(x,y,rscale = 1)
Freq = t.test(x,y,var.equal = T) %>% broom::tidy() %>% glue::glue_data("t({parameter}) = {round(statistic,2)}, p {ifelse(p.value < 0.001, '< .001', paste0('= ',substr(sprintf(p.value,fmt = '%.3f'),2,5)))}")
```

```r
tibble(x = seq(-4,4,0.001)) %>%
  mutate(posterior = dlogspline(x,fitPost),
         prior = dcauchy(x,0,1),
         likelihood = dnorm(x,-mean(zy),sd(zy)/sqrt(length(zy)))) %>%
  pivot_longer(cols = c(posterior,prior,likelihood), names_to = "type", values_to = "y") %>%
  ggplot(aes(x = x, y = y, colour = type, linetype = type)) + geom_line() +
  geom_point(data = . %>% filter(x == 0, type != "likelihood"),
             aes(x = x, y = y, colour = type),
             shape = 21, fill = "white", size = 4, show.legend = FALSE)  + 
  scale_color_manual(values = c(posterior = "red",prior = "black", likelihood = "blue"), name = NULL)  +
  scale_linetype_manual(values = c(posterior = 1,prior = 2, likelihood = 2),guide = FALSE) +
  theme_minimal() + labs(title = "Prior and Posterior",
                         subtitle = glue::glue("JSZ BF10: {round(exp(BF_pkg@bayesFactor$bf),2)}
                                               Savage-Dickey Method: {round(BF_SD,2)}
                                               t-test (equal var): {Freq}"),
                         x = "Effect (δ)",
                         y = "Density") +
  theme(legend.position = "top") 
```
