# Choosing priors Part I

```r
require(tidyverse)
require(patchwork)
```

```r
# 1

# plot a beta prior with a = b = 'a very small number'

a = 1e-6
b = 1e-6

tibble(theta = seq(0,1, length.out = 100000)) %>%
mutate(y = dbeta(x = theta, shape1 = a, shape2 = b)) %>%
ggplot(aes(x = theta, y = y)) + geom_line() + theme_minimal() + 
scale_x_continuous(name = "θ", breaks = seq(0,1,.2)) +
scale_y_continuous(name = "p(θ)")
```

```r
# 2

# generate all the plots

plots = map(c(1,50,99), function(x)
    tibble(theta = seq(0,1,.001),
           like = dbinom(x = x, size = 100, prob = seq(0,1,.001))) %>% 
    ggplot(aes(x = theta, y = like)) + geom_line() + 
    scale_x_continuous(name = "θ", breaks = seq(0,1,.2)) + 
    scale_y_continuous(name = "L(θ") + theme_minimal() + 
    labs(title = glue("{x} heads in 100 flips"))) 


wrap_plots(plots, nrow = 3) + plot_annotation(tag_levels = "A")
```
